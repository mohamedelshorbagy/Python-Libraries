# Seaborn Library

```python
print("introduction to seaborn")
```

```python
import seaborn as sns
```

```python
import numpy as np
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

# DistPlot

```python
# First Plot Type ( DistPlot )
```

```python
#distplot( dataset[colName] , kdeValue[True , False] , bins[Number refer to the number of bins]  )
```

```python
sns.distplot(tips['total_bill'] , kde=False , bins=60)
```

# jointPlot

```python
# Second Plot Type ( JointPlot )
```

```python
# jointplot( x = 'colName'  , y = 'colName' , data=dataset , kind ['kde' , 'hex' , 'reg' , 'scatter' , ..etc[Note : scatter Plot is Default]] ) 
```

```python
sns.jointplot(x ='size' , y ='total_bill' ,data = tips , kind='hex')
```

# PairPlot

```python
# Third Plot Type ( PairPlot ) 
```

```python
# Note : 
    # Only Works and PLots the Number Cols
```

```python
# pairplot( dataset , hue=[col with string that have two or more values which are different ] , palette=['husl' , 'coolwarm']{style the hue plot} )    
```

```python
sns.pairplot(tips , hue="sex" , palette="coolwarm")
```

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill'])
```

```python
# --------------------------------- End of Distribution Curves Lecture ----------------------------------------------- #
```

# Start Categorial Plots

```python
# First type Plot ( BoxPlot )
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='sex',palette='husl')
```

```python
# Second Plot ( count Plot )
```

```python
sns.countplot(x='sex',data=tips)
```

```python
#Third Plot ( violinplot )
```

```python
sns.violinplot(x = 'day',y='total_bill' , data=tips , hue='sex' , palette='husl', split=True)
```

```python
#Fourth Plot ( bars )
```

```python
sns.barplot(x='day',y='total_bill',data=tips)
```

```python
#Fifth Plot ( factorPlot )
```

```python
sns.factorplot(x = 'day', y = 'total_bill' , data=tips )
```

```python
# Six Plot ( stripPlot )
```

```python
sns.stripplot(x = 'day' , y='total_bill', data=tips)
```
