```python
x = np.linspace(0 ,10 , 10)
```

```python
y = x ** 2
```

```python
# Plotting the Prabola 
```

```python
#Functional Method
plt.plot(x , y , 'r')
plt.title('Prabola Function')
plt.xlabel('X Lable')
plt.ylabel('Y Lable')
```

```python

# Object Oriented Method
fig = plt.figure()   # Creat the Canvas

# It takes Four Parameters-> add_axes ===> Left , top , Height , Width  ---> Values From 0 -> 1 
axes = fig.add_axes([0.1,0.1,0.8,0.8])


# Plot Two Plots in the Axes 
axes.plot(x , y , label='X Squared')
axes.plot(y , x,label='X CUbe')


# Put Titles in the Plotting Area
axes.set_title('Big Plot')

# Put the Labels in the Axes
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')

axes.legend(loc=0)
```

# Figure Subplots in the Same Canvas

```python
fig,axes = plt.subplots(3,3)

# Note : Axes is a 2D Numpy Array 
for ax in axes:   # For the First Array ==> First Index
    for figures in ax: # For the Second Index === Second Array 
        figures.plot(x,y,label='X Squared')
        figures.set_title('Tit')
        figures.set_xlabel('X Squred')
        figures.set_ylabel('Y')
        
        

plt.tight_layout()
```

# Styling the MatplotLib

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x , x**3  ,label='X ^ 3')
axes.plot(x , x**4 , label='X ^ 4')
axes.plot(x , x**2 , label='X ^ 2')
axes.plot(x, x +3500, label='X ^ 1')
axes.legend(loc=0)
```

# Using Subplots 

```python
fig , axes = plt.subplots(1 , 2)

# By Default ake the Add_axes for the Canvases

for ax in axes:
    ax.plot(x , y ,'y' ,label='x',marker='o')   # label Refer to the Legend Text String 
    ax.legend(loc=0)  # 0 ===> Refers to the Best Location in the Canvas 

plt.tight_layout()

```
